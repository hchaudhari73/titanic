```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
def age_grp(age):
    if age<2:
        return "infant"
    elif age<18:
        return "child"
    elif age<60:
        return "adult"
    else:
        return "senior"

titanic["age_group"] = titanic.Age.map(age_grp)
```

```python
titanic.head(15)
```

```python
titanic["Fare_group"] = pd.qcut(titanic.Fare,3, labels = ["low","medium","high"])
titanic.head(15)
```

```python
fare_sur = titanic.groupby("Fare_group")[["Survived"]].count()
```

```python
df = round(titanic.groupby(["age_group","Fare_group","Survived"])[["Survived"]].count()/titanic.groupby(["age_group","Fare_group"])[["Survived"]].count()*100,2)
df.rename(columns = {"Survived":"Survived_per"})
```

```python
print(titanic.groupby(["age_group","Fare_group"])[["Survived"]].count())
titanic.groupby(["age_group","Fare_group"])[["Survived"]].count().plot(kind = "bar")
plt.show()
```
